```python
import os

from scanpy import read_h5ad

dataset_path = (
    "/dccstor/bmfm-targets/data/omics/transcriptome/scRNA/pretrain/cellxgene/"
)

anndata_files = "anndata_all/all_unique_cells/"
files_path = dataset_path + anndata_files


all_files = os.listdir(files_path)
all_files.sort()

def get_unique_counts(df, result_dict):
    for column in df.select_dtypes(include='category').columns:
        counts = df[column].value_counts().to_dict()
        if column in result_dict:
            for value, count in counts.items():
                result_dict[column][value] = result_dict[column].get(value, 0) + count
        else:
            result_dict[column] = counts

    return result_dict

result_dict = {}

# Loop through dataframes and update the dictionary
for i, df in enumerate(all_files):
    current_file = read_h5ad(files_path + df)
    result_dict = get_unique_counts(current_file.obs, result_dict)
```

The cellXgene holds multiple metadata fields. Most of them have a string field, and a parallel "ontolgy term ID" field. The tissue field has a general and broader categories field. 
Here are the list of all the metadata fields: 

```python
sorted(result_dict.keys())
```

Next are printed the different metadata fields (except for the ontology term IDs and the dataset and donor IDS), and the different values distribution: 

```python
ignore_keys = {'dataset_id', 'donor_id', 'assay_ontology_term_id', 'cell_type_ontology_term_id', 'development_stage_ontology_term_id', 'disease_ontology_term_id', 
               'self_reported_ethnicity_ontology_term_id', 'sex_ontology_term_id', 'tissue_general_ontology_term_id', 'tissue_ontology_term_id', }
filtered_result_dict = {k:v for k,v in result_dict.items() if k not in ignore_keys}
pprint.pprint(filtered_result_dict)
```

```python
special_tokens = ["CELL_TYPE_CLASS", "TISSUE_TYPE_CLASS", "MOLECULAR_ENTITY_CELL_GENE_EXPRESSION_RANKED"]
output_file = "cell_type_vocabulary.txt"
with open(output_file, "w") as f:
    f.write("#Cellxgene cell types ontology term id\n")
    for cell_type in result_dict['cell_type_ontology_term_id'].keys():
        f.write(cell_type+'\n')
    
    f.write("#Cellxgene tissue ontology term id\n")
    for cell_type in result_dict['tissue_ontology_term_id'].keys():
        f.write(cell_type+'\n')
    
    f.write("#Special tokens\n")
    for token in special_tokens:
        f.write(token+"\n")

    
```
